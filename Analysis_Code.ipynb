```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load in your dataset
airlinefile = '/british airline.csv'
df = pd.read_csv(airlinefile)

economy_class_df = df[df['Seat Type'] == 'Economy Class']
value_counts = economy_class_df['Value For Money'].value_counts()

# Sort the ratings and their corresponding counts in ascending order
value_counts_sorted = value_counts.sort_index()

# Define colors for each section
colors = ['#8a2be2', '#da70d6', '#ff69b4', '#ff1493', '#00bfff']

# Define labels for each rating and sort based on value counts
stars = ['1 star', '2 stars', '3 stars', '4 stars', '5 stars']
stars_sorted = [stars[i-1] for i in value_counts_sorted.index]

# Add visual descriptors and styles
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
plt.pie(value_counts_sorted, labels=stars_sorted, colors=colors, autopct='%1.1f%%')
plt.title('Value for Money: Economy Class (1 - 5 stars)')
plt.axis('equal')

# Create custom legend based on sorted value counts and colors
custom_legend = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
plt.legend(custom_legend, stars_sorted, title='Value Ratings', loc='center left', bbox_to_anchor=(1, 0.5))

# Show the pie chart
plt.show()
```

BAR CHART COMPARING CLASS RATING AND IF THEY WOULD FLY AGAIN


```python
#BAR CHART COMPARING CLASS RATING AND IF THEY WOULD FLY AGAIN
import pandas as pd
import matplotlib.pyplot as plt

# Reading data from our csv file
airline ='/british airline.csv'
df = pd.read_csv(airline)

# Group data by Seat Type and Recommended
group = df.groupby(['Seat Type', 'Recommended']).size().unstack()
seat_type_order = ['Economy Class', 'Premium Economy','Business Class','First Class']

# Reorder the rows
group = group.reindex(seat_type_order)

# Plot the bar chart
group.plot(kind='bar', color=['purple', 'pink'])

# Add labels and title
plt.xlabel('Seat Type')
plt.ylabel('Count')
plt.title('Responses for Different Ticket Status')

#label format
plt.xticks(rotation=25, ha='right')

# Show the plot
plt.legend(title='Recommended', loc='upper right')
plt.tight_layout()
plt.show()
```

Seat Type vs. cabin Service

```python
import pandas as pd  # import pandas
import matplotlib.pyplot as plt  # import matplotlib for plotting
from matplotlib.ticker import FuncFormatter  # import FuncFormatter for custom formatting

# read the BA dataset from the file path
airline = '/british airline.csv'
df = pd.read_csv(airline)

# filter dataset only for Economy Class rows
economy_class_data = df[df['Seat Type'] == 'Economy Class']

# change 'Cabin Staff Service' to numeric , coerce errors to non-numerical
economy_class_data['Cabin Staff Service'] = pd.to_numeric(economy_class_data['Cabin Staff Service'], errors='coerce')

# drop rows with non-numerical values in Cabin Staff Service column
economy_class_data.dropna(subset=['Cabin Staff Service'], inplace=True)

# count values for each rating in 'Cabin Staff Service' column
ratings_counts = economy_class_data['Cabin Staff Service'].value_counts()

# calculate total number of passengers in the filtered dataset
total_passengers = len(economy_class_data)

# calculate percentages of each rating
percentage_passengers = (ratings_counts / total_passengers) * 100

# make bar plot with % of passengers on y-axis and ratings on x-axis
plt.bar(percentage_passengers.index, percentage_passengers, color='pink', edgecolor='black')

# name each rating on the x-axis
plt.xticks(range(1, 6), ['1', '2', '3', '4', '5'])

# plot title and x/y-axis labels
plt.title('Cabin Staff Service Ratings in Economy Class')
plt.xlabel('Cabin Staff Service')
plt.ylabel('Percentage of Passengers')

# add slightly transparent grid lines for y-axis
plt.grid(axis='y', alpha=0.75)

# define and apply custom formatter to show y-axis ticks with the % sign
formatter = FuncFormatter(lambda x, _: f'{x:.0f}%')
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

Seat Comfort vs. Seat Type

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

airlinefile = '/british airline.csv'
df = pd.read_csv(airlinefile)

#pull first class data
econ_class_data = df[df['Seat Type'] == 'First Class']
econ_class_data['Seat Comfort'] = pd.to_numeric(econ_class_data['Seat Comfort'], errors='coerce')
econ_class_data.dropna(subset=['Seat Comfort'], inplace=True)

ratings_counts = econ_class_data['Seat Comfort'].value_counts()
total_passengers = len(econ_class_data)
percentage_passengers = (ratings_counts / total_passengers) * 100

plt.bar(percentage_passengers.index, percentage_passengers, color='fuchsia', edgecolor='black')
plt.xticks(range(1, 6), ['1', '2', '3', '4', '5'])

plt.title('Seat Comfort Ratings: First Class')
plt.xlabel('Seat Comfortability (Ranked 1-5)')
plt.ylabel('Percentage of Passengers')
plt.grid(axis='y', alpha=0.75)

formatter = FuncFormatter(lambda x, _: f'{x:.0f}%')
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

Value for Money (based on class type)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
airlinefile = '/british airline.csv'
df = pd.read_csv(airlinefile)

# Count the number of ratings for the  Economy class
premium_economy_df = df[df['Seat Type'] == 'Economy Class']
rating_counts = premium_economy_df['Rating'].value_counts()

# colors for each slice
colors = ['LightPink', 'HotPink', 'DeepPink', 'MediumPurple', 'PaleVioletRed', 'RoyalBlue', 'Turquoise', 'Magenta', 'BlueViolet', 'DarkSlateBlue']

# make sure colors and counts match based on rating counts
rating_counts_sorted = rating_counts.sort_index()
custom_colors_sorted = [colors[i-1] for i in rating_counts_sorted.index]

# Adding descriptions
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
plt.pie(rating_counts_sorted, labels=rating_counts_sorted.index, colors=custom_colors_sorted, autopct='%1.1f%%')  # Use index as labels
plt.title('Value for Money: Economy Class (Rating 1 - 10)')
plt.axis('equal')

# Created legend based on rating counts that match colors
legend = [plt.Rectangle((0,0),1,1, color=color) for color in custom_colors_sorted]
plt.legend(legend, rating_counts_sorted.index, title='Value Ratings', loc='center left', bbox_to_anchor=(1, 0.5))

# print pie chart
plt.show()
```
